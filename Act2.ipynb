```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .xlsx sin indice
data = pd.read_excel('Inmuebles2.xlsx')
```

```python
#Convertir DataFrame a CSV
data.to_csv("Inmuebles2.csv")
data
```

```python
#Eliminar primera fila o filas innecesarias
#Se elimina la primera fila que contenia información y solo nos quedamos con las etiquetas de cada columna
data= pd.read_csv('Inmuebles2.csv', skiprows=1, low_memory=False)
data
```

```python
#Eliminamos las columnas inecesarias que contienen los mismos datos "0" y "Referencias"
#Quedamos con 8 columnas
data= data.drop(['0','Referencia'], axis=1)
data
```

```python
#Identificar valores nulos
#Vemos que tenemos valores nulos en todas nuestras columnas
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Se copia el Dataframe
data2=data.copy()
```

```python
#Sustituir valores nulos con promedio o media (Precio Venta)
data2["Precio Venta"]=data2["Precio Venta"].fillna(round(data["Precio Venta"].mean(),1))
data2
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con mediana
#Superficie
data2["Superficie"]=data2["Superficie"].fillna(round(data["Superficie"].median(),1))
data2
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos hacia adelante
#Filtro por columnas

data2['Fecha Alta'] = data2['Fecha Alta'].fillna(method="ffill")
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos hacia atras
data2['Provincia'] = data2['Provincia'].fillna(method="bfill")
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Sustituir por un dato concreto
data2['Tipo'] = data2['Tipo'].fillna("Piso")  
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```
